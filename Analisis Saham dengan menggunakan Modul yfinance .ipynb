```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
```

```python
stock = ['AAPL', 'AMZN', 'TSLA', 'MSFT', 'INTC', 'GOOGL']
stocks = yf.download(stock, start = '2017-01-01', end = '2021-07-29')
data = stocks.loc[:, "Close"].copy()
```

```python
# Build stock chart using python

data.plot(figsize=(17, 8), fontsize= 18)
plt.style.use('seaborn')
plt.show()
```

```python
# Measure stocks performance

normdata = data.div(data.iloc[0]).mul(100)
normdata.plot(figsize=(17, 8), fontsize = 18)
plt.style.use('seaborn')
plt.show()
```

```python
# Stock performance frequency

google = normdata.GOOGL.copy().to_frame()
ret = google.pct_change().dropna()

ret.plot(kind='hist', figsize=(12, 8), bins=100)
plt.show()
```

```python
tesla = normdata.TSLA.copy().to_frame()
ret = tesla.pct_change().dropna()

ret.plot(kind='hist', figsize=(12, 8), bins=100)
plt.show()
```

```python
apple = normdata.AAPL.copy().to_frame()
ret = apple.pct_change().dropna()

ret.plot(kind='hist', figsize=(12, 8), bins=100)
plt.show()
```

```python
#Stock risk analysis

data.pct_change()
```

```python
com = data.pct_change().dropna()
com.describe()
```

```python
sum = com.describe().T.loc[:, ['mean', 'std']]
sum
```

```python
sum['mean'] = sum['mean']*252
sum['std'] = sum['std']* np.sqrt(252)
sum
```

```python
sum.plot.scatter(x = 'std', y = 'mean', figsize =(20,8), s=50, fontsize=15)
for i in sum.index:
    plt.annotate(i, xy=(sum.loc[i, 'std'] + 0.002, sum.loc[i, 'mean'] + 0.002), size=15)
```
